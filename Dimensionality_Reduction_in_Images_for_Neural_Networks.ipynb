Bibliotecas necessárias

```python
from PIL import Image
import matplotlib.pyplot as plt
```

Função para converter imagem colorida para escala de cinza

```python
def converter_para_cinza(imagem_colorida):
    largura, altura = imagem_colorida.size
# Criar uma nova imagem em tons de cinza
    imagem_cinza = Image.new('L', (largura, altura))

    for x in range(largura):
        for y in range(altura):
# Obter valores RGB
            r, g, b = imagem_colorida.getpixel((x, y))
# Fórmula de conversão para cinza (luminância)
            intensidade_cinza = int(0.299 * r + 0.587 * g + 0.114 * b)
            imagem_cinza.putpixel((x, y), intensidade_cinza)

    return imagem_cinza
```

Função para binalizar uma imagem em tons de cinza

```python
def binarizar_imagem(imagem_cinza, limiar):
    largura, altura = imagem_cinza.size
# Criar uma nova imagem binária
    imagem_binaria = Image.new('1', (largura, altura))

    for x in range(largura):
        for y in range(altura):
            # Obter valor do pixel
            intensidade = imagem_cinza.getpixel((x, y))
            # Comparar com o limiar
            imagem_binaria.putpixel((x, y), 1 if intensidade > limiar else 0)

    return imagem_binaria

# Carregar imagem colorida
imagem_colorida = Image.open('imagem_colorida.png')

# Converter para escala de cinza
imagem_cinza = converter_para_cinza(imagem_colorida)

# Salvar a imagem em cinza
imagem_cinza.save('imagem_cinza.png')

# Binarizar a imagem
limiar = 127  # Definir limiar para binarização
imagem_binaria = binarizar_imagem(imagem_cinza, limiar)
# Salvar a imagem binária
imagem_binaria.save('imagem_binaria.png')

# Exibir imagem em tons de cinza
imagem_cinza.show()
# Exibir imagem binária
imagem_binaria.show()
```

Código para exibir a imagem original (colorida)

```python
plt.figure(figsize=(8, 8))
plt.title("Imagem Colorida")
plt.imshow(imagem_colorida)
plt.axis("off")
plt.show()
```

Código para exibir a imagem gerada em tons de cinza

```python
plt.figure(figsize=(8, 8))
plt.title("Imagem em Tons de Cinza")
plt.imshow(imagem_cinza, cmap="gray")
plt.axis("off")
plt.show()
```

Código para exibir a imagem binarizada

```python
plt.figure(figsize=(8, 8))
plt.title("Imagem Binarizada")
plt.imshow(imagem_binaria, cmap="gray")
plt.axis("off")
plt.show()
```

**Conclusão**

No momento da binalização, foi escolhido o limiar 127 pois este valor é simples e eficaz, eliminando a necesidade de cálculos adicionais ou análise complexa. Ademais, se houvesse escolhido valores muito baixo, isso geraria imagens com mais áreas brancas e ao contrário (valores mais altos de limiar) geraria áreas mais pretas.Portanto, o limiar 127 foi escolhido como um valor padrão seguro para separar tons claros e escuros.
